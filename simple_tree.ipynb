# Import libraries

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import joblib
import matplotlib.pyplot as plt
```

# Load the dataset

```python
# read the csv file
df = pd.read_csv("students_senior_dataset.csv")

# show first rows
df.head()
```

```python
df.info()
```

# Simple preprocessing

```python
# convert Yes/No to numbers
df = df.replace({"Yes": 1, "No": 0})

# separate features and target
X = df.drop("Senior", axis=1)
y = df["Senior"]
```

# Train + visualize the decision tree

```python
# create and train model
model = DecisionTreeClassifier()
model.fit(X, y)

# plot the tree
plt.figure(figsize=(18,10))
plot_tree(
    model,
    feature_names=X.columns,
    class_names=["No", "Yes"],
    filled=True
)

plt.show()
```

# Save the model

```python
# save the model
joblib.dump(model, "senior_classifier_model.pkl")

print("Model saved")
```
